```python
# Q2. Data Cleaning and Preparation
# Identify missing values using isnull and isna.
# Handle missing values using fillna, dropna, interpolation.
# Apply scaling techniques (min-max, z-score) to numerical columns.
# Create dummy variables for categorical columns.

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df_csv = pd.read_csv(r"C:\Users\AxD\OneDrive\Desktop\inevitable\predictive lab code\small_sales_data.csv")

print("Missing values in each column:")
print(df_csv.isnull().sum())
print()

df_fillna = df_csv.fillna(0)
print("After filling missing values with 0:")
print(df_fillna)
print()

df_interpolated = df_csv.interpolate()
print("After interpolating missing values:")
print(df_interpolated)
print()

numerical_columns = ['Units Sold', 'Unit Price', 'Total Sales']
scaler = MinMaxScaler()
df_minmax_scaled = df_interpolated.copy()
df_minmax_scaled[numerical_columns] = scaler.fit_transform(df_interpolated[numerical_columns])
print("After Min-Max Scaling:")
print(df_minmax_scaled)
print()

scaler_z = StandardScaler()
df_z_scaled = df_interpolated.copy()
df_z_scaled[numerical_columns] = scaler_z.fit_transform(df_interpolated[numerical_columns])
print("After Z-score Scaling:")
print(df_z_scaled)
print()

df_with_dummies = pd.get_dummies(df_interpolated, columns=['Product No'])
print("After creating dummy variables:")
print(df_with_dummies)

```
